```python
# tables to work with for temporary storage of generation and consumption generationTable and consumptionTable
# 1. For each windmill, we allocate the generation to the customers in the order of priority for each slot
# 2. If generation for current slot and windmill is greater than consumption of current slot and customer, allot consumption and add the data to AllocationTable, subtract consumption from consumptionTable and subtract consumption from generationTable 
# 3. If Consumption greater current slot and customer than generation of current slot and windmill, allot generation and add the data to AllocationTable and subtract generation from generationTable
# 4. If generation of current slot and windmill or consumption of current slot and customer is 0, skip the customer or windmill
# 5. Repeat process until generation=0 or consumption=0 for all customers or windmills
```

```python

df['735']=[0 if x==0 else 1 for x in df['735']]
df['736']=[0 if x==0 else 1 for x in df['736']]
df['897']=[0 if x==0 else 1 for x in df['897']]
df['512']=[0 if x==0 else 1 for x in df['512']]
df['D042']=[0 if x==0 else 1 for x in df['D042']]
df['D228']=[0 if x==0 else 1 for x in df['D228']]
df['739']=[0 if x==0 else 1 for x in df['739']]
# print(df.info())

def getSlotPriority(inputSortedDf,slotInfo):
    df1=inputSortedDf.copy()
    df1.sort_values(by=slotInfo,ascending=False,inplace=True)
    # print(df1)
    # df[slotInfo]=float(df[slotInfo])
    counter=1
    df1[slotInfo].reset_index(drop=True, inplace=True)
    previousRate=df1[slotInfo][0]
    priorityList=[1,]
    # df['C1/C2 Priority'][0]=1
    for x in range(1,len(df1)):
        # print(previousRate,df1[slotInfo][x])
        if previousRate==df1[slotInfo][x]:
            priorityList.append(counter)
            previousRate=df1[slotInfo][x]
        else:
            counter=counter+1
            priorityList.append(counter)
            previousRate=df1[slotInfo][x]
    df1[slotInfo+' Priority']=priorityList
    return df1
c1SortedPriority=pd.DataFrame(getSlotPriority(df.sort_values(by='C1/C2',ascending=False),'C1/C2'))
# print(c1SortedPriority)
c4SortedPriority=getSlotPriority(c1SortedPriority.sort_values(by='C4',ascending=False),'C4')
# print(c4SortedPriority)
c5SortedPriority=getSlotPriority(c4SortedPriority.sort_values(by='C5',ascending=False),'C5')
print(c5SortedPriority)
dfx=c5SortedPriority.copy()
dfx['C1/C2 Priority/735']=dfx['735']*dfx['C1/C2 Priority']
dfx['C1/C2 Priority/736']=dfx['736']*dfx['C1/C2 Priority']
dfx['C1/C2 Priority/897']=dfx['897']*dfx['C1/C2 Priority']
dfx['C1/C2 Priority/512']=dfx['512']*dfx['C1/C2 Priority']
dfx['C1/C2 Priority/D042']=dfx['D042']*dfx['C1/C2 Priority']
dfx['C1/C2 Priority/D228']=dfx['D228']*dfx['C1/C2 Priority']
dfx['C1/C2 Priority/739']=dfx['739']*dfx['C1/C2 Priority']
dfx['C4 Priority/735']=dfx['735']*dfx['C4 Priority']
dfx['C4 Priority/736']=dfx['736']*dfx['C4 Priority']
dfx['C4 Priority/897']=dfx['897']*dfx['C4 Priority']
dfx['C4 Priority/512']=dfx['512']*dfx['C4 Priority']
dfx['C4 Priority/D042']=dfx['D042']*dfx['C4 Priority']
dfx['C4 Priority/D228']=dfx['D228']*dfx['C4 Priority']
dfx['C4 Priority/739']=dfx['739']*dfx['C4 Priority']
dfx['C5 Priority/735']=dfx['735']*dfx['C5 Priority']
dfx['C5 Priority/736']=dfx['736']*dfx['C5 Priority']
dfx['C5 Priority/897']=dfx['897']*dfx['C5 Priority']
dfx['C5 Priority/512']=dfx['512']*dfx['C5 Priority']
dfx['C5 Priority/D042']=dfx['D042']*dfx['C5 Priority']
dfx['C5 Priority/D228']=dfx['D228']*dfx['C5 Priority']
dfx['C5 Priority/739']=dfx['739']*dfx['C5 Priority']

# dfx['736']=dfx['736']*dfx['C1/C2 Priority']
print(dfx)

            
  
    
    
```

```python
slots=['C1/C2','C4','C5']
allWindmills=dfWindmills['Windmill'].tolist()
allCustomers=dfConsumption['Customer'].tolist()
generationTable.to_excel('generationTable.xlsx')
consumptionTable.to_excel('consumptionTable.xlsx')

for currWindmill in allWindmills:
    
    currAllocationTable={}
    for slot in slots:
        # create data structure with priority, customer and consumption
        columnToCheck=slot+" Priority/"+currWindmill
        dfSortedCustomer=dfx.sort_values(by=columnToCheck).copy()
        dfSortedCustomer=dfSortedCustomer[dfSortedCustomer[columnToCheck]>0]
        uniquePriorities=dfSortedCustomer[columnToCheck].unique().tolist()
        currSlotGen=generationTable[generationTable['Windmill']==currWindmill][slot].values[0]
        for i in uniquePriorities:
            currAllocationTable[i]={'Customer':[],'Consumption':[]}
        for index,row in dfSortedCustomer.iterrows():
            currAllocationTable[row[columnToCheck]]['Customer'].append(row['Customer'])
            currAllocationTable[row[columnToCheck]]['Consumption'].append(consumptionTable[consumptionTable['Customer']==row['Customer']][slot].values[0])
        for i in uniquePriorities:
            if currSlotGen<=0:
                currAllocationTable={}
                break
            
            if len(currAllocationTable[i]['Customer'])==1:
                if currAllocationTable[i]['Consumption'][0]<=0:
                    continue
                #allocate
                if currSlotGen>=currAllocationTable[i]['Consumption'][0]:
                    allocationTable['Windmill'].append(currWindmill)
                    allocationTable['Customer'].append(currAllocationTable[i]['Customer'][0])
                    allocationTable[slot].append(currAllocationTable[i]['Consumption'][0])
                    for x in slots:
                        if x!=slot:
                            allocationTable[x].append(0)
                    currSlotGen=currSlotGen-currAllocationTable[i]['Consumption'][0]
                    currAllocationTable[i]['Consumption'][0]=0
                    generationTable.loc[generationTable['Windmill']==currWindmill,slot]=currSlotGen
                    consumptionTable.loc[consumptionTable['Customer']==currAllocationTable[i]['Customer'][0],slot]=currAllocationTable[i]['Consumption'][0]
                else:
                    allocationTable['Windmill'].append(currWindmill)
                    allocationTable['Customer'].append(currAllocationTable[i]['Customer'][0])
                    allocationTable[slot].append(currSlotGen)
                    for x in slots:
                        if x!=slot:
                            allocationTable[x].append(0)
                    currAllocationTable[i]['Consumption'][0]=currAllocationTable[i]['Consumption'][0]-currSlotGen
                    currSlotGen=0
                    generationTable.loc[generationTable['Windmill']==currWindmill,slot]=currSlotGen
                    consumptionTable.loc[consumptionTable['Customer']==currAllocationTable[i]['Customer'][0],slot]=currAllocationTable[i]['Consumption'][0]
            else:
                #split and allocate
                dfCustomersOfCurrentPriority=pd.DataFrame(currAllocationTable[i])
                #filter only non zero consumption
                
                while currSlotGen>1:
                    print(currSlotGen,len(dfCustomersOfCurrentPriority))
                    if len(dfCustomersOfCurrentPriority)==0:
                        break
                    dfCustomersOfCurrentPriority=dfCustomersOfCurrentPriority[dfCustomersOfCurrentPriority['Consumption']>0]
                    
                    # Sort by consumption
                    dfCustomersOfCurrentPriority=dfCustomersOfCurrentPriority.sort_values(by='Consumption',ascending=False)
                    # divide available power by total customers
                    currSlotGenPerCustomer=currSlotGen/len(dfCustomersOfCurrentPriority)
                    # allocate power to each customer with condition that power is less than consumption or greater than consumption
                    
                    for index,row in dfCustomersOfCurrentPriority.iterrows():
                        if currSlotGenPerCustomer<=0:
                            break
                        if row['Consumption']<=currSlotGenPerCustomer:
                            allocationTable['Windmill'].append(currWindmill)
                            allocationTable['Customer'].append(row['Customer'])
                            allocationTable[slot].append(row['Consumption'])
                            for x in slots:
                                if x!=slot:
                                    allocationTable[x].append(0)
                            currSlotGen=currSlotGen-row['Consumption']
                            currAllocationTable[i]['Consumption'][index]=0
                            generationTable.loc[generationTable['Windmill']==currWindmill,slot]=currSlotGen
                            consumptionTable.loc[consumptionTable['Customer']==row['Customer'],slot]=currAllocationTable[i]['Consumption'][index]
                            # update the dfCustomersOfCurrentPriority with remaining consumption
                            dfCustomersOfCurrentPriority.loc[index,'Consumption']=0
                        else:
                            allocationTable['Windmill'].append(currWindmill)
                            allocationTable['Customer'].append(row['Customer'])
                            allocationTable[slot].append(currSlotGenPerCustomer)
                            for x in slots:
                                if x!=slot:
                                    allocationTable[x].append(0)
                            currAllocationTable[i]['Consumption'][index]=currAllocationTable[i]['Consumption'][index]-currSlotGenPerCustomer
                            currSlotGen=currSlotGen-currSlotGenPerCustomer
                            generationTable.loc[generationTable['Windmill']==currWindmill,slot]=currSlotGen
                            consumptionTable.loc[consumptionTable['Customer']==row['Customer'],slot]=currAllocationTable[i]['Consumption'][index]
                            # update the dfCustomersOfCurrentPriority with remaining consumption
                            dfCustomersOfCurrentPriority.loc[index,'Consumption']=currAllocationTable[i]['Consumption'][index]
dfAllocation=pd.DataFrame(allocationTable)
dfAllocation.to_excel('allocationTable.xlsx')
```
